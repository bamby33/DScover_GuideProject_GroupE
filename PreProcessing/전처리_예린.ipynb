```python
import pandas as pd
import numpy as np

train = pd.read_csv("train.csv")
sanji = pd.read_csv("TRAIN_산지공판장_2018-2021.csv")
wholesale = pd.read_csv("TRAIN_전국도매_2018-2021.csv")
```

```python
for df in [sanji, wholesale]:
    df["품목명"] = df["품목명"].astype(str).str.strip()
    df["품종명"] = df["품종명"].astype(str).str.strip()

if "등급명" in sanji.columns:
    sanji["등급명"] = sanji["등급명"].astype(str).str.strip()

# =========================
# 3. 타겟 정의 (리스트 형태로 통일)
# =========================
target_dict = {
    "건고추": ["화건"],
    "사과": ["홍로", "후지"],
    "감자": ["수미"],
    "배": ["신고"],
    "마늘": ["깐마늘"],   # 필요시 수정 가능
    "무": ["무"],
    "상추": ["청상추"],
    "배추": ["배추"],
    "양파": ["양파"],
    "대파": ["대파(일반)"]
}

# 등급 조건 (산지 전용)
grade_dict = {
    "건고추": "상",
    "사과": "상",
    "감자": "상",
    "배": "상",
    "마늘": "상",
    "무": "상",
    "상추": "상",
    "배추": "상",
    "양파": "상",
    "대파": "상"
}

# =========================
# 4. 산지공판장 필터링 (등급 포함)
# =========================
sanji_list = []

for item in target_dict:
    temp = sanji[
        (sanji["품목명"] == item) &
        (sanji["품종명"].isin(target_dict[item])) &
        (sanji["등급명"] == grade_dict[item])
    ]
    sanji_list.append(temp)

sanji_target = pd.concat(sanji_list, ignore_index=True)

# =========================
# 5. 전국도매 필터링 (등급 없음)
# =========================
wholesale_list = []

for item in target_dict:
    temp = wholesale[
        (wholesale["품목명"] == item) &
        (wholesale["품종명"].isin(target_dict[item]))
    ]
    wholesale_list.append(temp)

wholesale_target = pd.concat(wholesale_list, ignore_index=True)

# =========================
# 6. 결과 확인
# =========================
print("산지 데이터:")
print(sanji_target["품목명"].value_counts())

print("\n도매 데이터:")
print(wholesale_target["품목명"].value_counts())

wholesale_target.head()
```

```python
import pandas as pd
import numpy as np

# =========================
# 1. 품종명 정규화 함수
# =========================
def normalize_variety(item, variety):
    v = str(variety).strip().replace(' ', '')

    if item == '감자':
        mapping = {
            '감자수미': '수미',
            '감자수미(햇)': '수미',
            '감자수미(저장)': '수미',
            '감자대지': '대지',
            '감자두백': '두백',
            '감자조풍': '조풍',
            '감자수입': '감자(수입)',
            '감자': '기타감자',
            '홍감자': '홍감자',
        }
        return mapping.get(v, v)

    if item == '대파':
        mapping = {
            '대파(일반)': '대파(일반)',
            '대파수입': '대파(수입)',
            '실파': '기타대파',
            '깐쪽파': '깐대파',
            '쪽파': '중파',
        }
        return mapping.get(v, v)

    if item == '무':
        mapping = {
            '무': '무',
            '다발무': '다발무',
            '열무': '기타무',
        }
        return mapping.get(v, v)

    if item == '배':
        return v

    if item == '배추':
        mapping = {
            '배추': '배추',
            '쌈배추': '쌈배추',
            '절임배추': '절임배추',
            '알배기배추': '기타배추',
            '얼갈이배추': '기타배추',
            '봄동배추': '봄배추',
        }
        return mapping.get(v, v)

    if item == '사과':
        mapping = {
            '후지': '후지',
            '홍로': '홍로',
            '쓰가루': '아오리',
        }
        return mapping.get(v, v)

    if item == '상추':
        mapping = {
            '청': '청상추',
            '적': '적상추',
        }
        return mapping.get(v, v)

    if item == '양파':
        mapping = {
            '양파': '양파',
            '양파(햇)': '양파(일반)',
            '양파수입': '양파(수입)',
            '자주양파': '자주양파',
            '저장양파': '저장양파',
            '조생양파': '조생양파',
        }
        return mapping.get(v, v)

    # 건고추, 마늘 등은 매칭 어려움 → 품목 기준 fallback
    if item in ['건고추', '깐마늘(국산)']:
        return None

    return v


# =========================
# 2. 메인 함수
# =========================
def add_wholesale_volume(test_path, wholesale_path, save_path=None):

    test = pd.read_csv(test_path)
    wh = pd.read_csv(wholesale_path)

    # ----------------------
    # 시점 맞추기
    # ----------------------
    wh['시점'] = wh['시점'].replace({'T': 'T-0순'})

    # ----------------------
    # 품종명 정규화
    # ----------------------
    test['품종명_std'] = test.apply(
        lambda r: normalize_variety(r['품목명'], r['품종명']), axis=1
    )
    wh['품종명_std'] = wh.apply(
        lambda r: normalize_variety(r['품목명'], r['품종명']), axis=1
    )

    # ----------------------
    # 전국도매 집계
    # ----------------------
    # (1) 품종 기준
    wh_var = (
        wh.groupby(['시점','품목명','품종명_std'], dropna=False)['총반입량(kg)']
        .sum()
        .reset_index()
        .rename(columns={'총반입량(kg)': '도매_총반입량_품종'})
    )

    # (2) 품목 기준 fallback
    wh_item = (
        wh.groupby(['시점','품목명'])['총반입량(kg)']
        .sum()
        .reset_index()
        .rename(columns={'총반입량(kg)': '도매_총반입량_품목'})
    )

    # ----------------------
    # merge
    # ----------------------
    merged = test.merge(
        wh_var,
        on=['시점','품목명','품종명_std'],
        how='left'
    ).merge(
        wh_item,
        on=['시점','품목명'],
        how='left'
    )

    # fallback 적용
    merged['도매_총반입량'] = merged['도매_총반입량_품종'].fillna(
        merged['도매_총반입량_품목']
    )

    # ----------------------
    # 정리
    # ----------------------
    result = merged[test.columns.tolist() + ['도매_총반입량']].copy()

    # 저장 옵션
    if save_path:
        result.to_csv(save_path, index=False, encoding='utf-8-sig')

    print("매칭률:", round(result['도매_총반입량'].notna().mean()*100, 2), "%")

    return result

    test_files = ["TEST_00.csv", "TEST_01.csv", "TEST_02.csv","TEST_0.csv"]

for i, f in enumerate(test_files):
    add_wholesale_volume(
        test_path=f,
        wholesale_path=f.replace("TEST_", "TEST_전국도매_"),
        save_path=f"processed_{f}"
    )
```

```python
import pandas as pd
import numpy as np
import os

# =========================
# 1. 품종명 정규화 함수
# =========================
def normalize_variety(item, variety):
    v = str(variety).strip().replace(' ', '')

    if item == '감자':
        mapping = {
            '감자수미': '수미',
            '감자수미(햇)': '수미',
            '감자수미(저장)': '수미',
            '감자대지': '대지',
            '감자두백': '두백',
            '감자조풍': '조풍',
            '감자수입': '감자(수입)',
            '감자': '기타감자',
            '홍감자': '홍감자',
            '수미': '수미',
            '대지': '대지',
            '두백': '두백',
            '조풍': '조풍',
        }
        return mapping.get(v, v)

    if item == '대파':
        mapping = {
            '대파(일반)': '대파(일반)',
            '대파수입': '대파(수입)',
            '실파': '기타대파',
            '깐쪽파': '깐대파',
            '쪽파': '중파',
        }
        return mapping.get(v, v)

    if item == '무':
        mapping = {
            '무': '무',
            '다발무': '다발무',
            '열무': '기타무',
        }
        return mapping.get(v, v)

    if item == '배':
        mapping = {
            '신고': '신고',
            '원황': '원황',
        }
        return mapping.get(v, v)

    if item == '배추':
        mapping = {
            '배추': '배추',
            '쌈배추': '쌈배추',
            '절임배추': '절임배추',
            '알배기배추': '기타배추',
            '얼갈이배추': '기타배추',
            '봄동배추': '봄배추',
        }
        return mapping.get(v, v)

    if item == '사과':
        mapping = {
            '후지': '후지',
            '홍로': '홍로',
            '쓰가루': '아오리',
            '아오리': '아오리',
        }
        return mapping.get(v, v)

    if item == '상추':
        mapping = {
            '청': '청상추',
            '적': '적상추',
            '청상추': '청상추',
            '적상추': '적상추',
        }
        return mapping.get(v, v)

    if item == '양파':
        mapping = {
            '양파': '양파',
            '양파(햇)': '양파(일반)',
            '양파수입': '양파(수입)',
            '자주양파': '자주양파',
            '저장양파': '저장양파',
            '조생양파': '조생양파',
            '양파(일반)': '양파(일반)',
            '양파(수입)': '양파(수입)',
        }
        return mapping.get(v, v)

    # 매칭 어려운 품목은 품목 fallback용
    if item in ['건고추', '깐마늘(국산)']:
        return None

    return v


# =========================
# 2. 단일 파일 처리 함수
# =========================
def add_wholesale_volume(test_path, wholesale_path, save_path=None):
    test = pd.read_csv(test_path)
    wh = pd.read_csv(wholesale_path)

    # 전국도매 파일의 T를 test의 마지막 시점에 맞춤
    wh = wh.copy()
    wh['시점'] = wh['시점'].replace({'T': 'T-0순'})

    # 품종명 정규화
    test = test.copy()
    test['품종명_std'] = test.apply(
        lambda r: normalize_variety(r['품목명'], r['품종명']), axis=1
    )
    wh['품종명_std'] = wh.apply(
        lambda r: normalize_variety(r['품목명'], r['품종명']), axis=1
    )

    # 전국도매 집계
    wh_var = (
        wh.groupby(['시점', '품목명', '품종명_std'], dropna=False)['총반입량(kg)']
          .sum()
          .reset_index()
          .rename(columns={'총반입량(kg)': '도매_총반입량_품종'})
    )

    wh_item = (
        wh.groupby(['시점', '품목명'])['총반입량(kg)']
          .sum()
          .reset_index()
          .rename(columns={'총반입량(kg)': '도매_총반입량_품목'})
    )

    # merge
    merged = test.merge(
        wh_var,
        on=['시점', '품목명', '품종명_std'],
        how='left'
    ).merge(
        wh_item,
        on=['시점', '품목명'],
        how='left'
    )

    # fallback
    merged['도매_총반입량'] = merged['도매_총반입량_품종'].fillna(
        merged['도매_총반입량_품목']
    )

    # 원래 test 컬럼 + 새 컬럼만 남김
    result = merged[test.columns.tolist() + ['도매_총반입량']].copy()

    # 임시 컬럼 제거
    if '품종명_std' in result.columns:
        result = result.drop(columns=['품종명_std'])

    # 저장
    if save_path is not None:
        result.to_csv(save_path, index=False, encoding='utf-8-sig')

    match_rate = result['도매_총반입량'].notna().mean() * 100

    return result, match_rate


# =========================
# 3. TEST_00 ~ TEST_24 일괄 처리
# =========================
summary_list = []

for i in range(25):
    idx = f"{i:02d}"  # 00, 01, ..., 24

    test_path = f"TEST_{idx}.csv"
    wholesale_path = f"TEST_전국도매_{idx}.csv"
    save_path = f"TEST_{idx}_with_volume.csv"

    if not os.path.exists(test_path):
        print(f"[스킵] 파일 없음: {test_path}")
        continue

    if not os.path.exists(wholesale_path):
        print(f"[스킵] 파일 없음: {wholesale_path}")
        continue

    result, match_rate = add_wholesale_volume(
        test_path=test_path,
        wholesale_path=wholesale_path,
        save_path=save_path
    )

    summary_list.append({
        '파일명': f"TEST_{idx}",
        '행수': len(result),
        '채워진 행수': result['도매_총반입량'].notna().sum(),
        '결측 행수': result['도매_총반입량'].isna().sum(),
        '매칭률(%)': round(match_rate, 2)
    })

    print(f"[완료] TEST_{idx} -> {save_path} | 매칭률: {match_rate:.2f}%")


# =========================
# 4. 요약표 확인
# =========================
summary_df = pd.DataFrame(summary_list)
print("\n전체 요약")
print(summary_df)

summary_df.to_csv("wholesale_matching_summary.csv", index=False, encoding='utf-8-sig')
```
